```python
import numpy as np
import skfuzzy as fuzzy
import matplotlib.pyplot as plt

white=np.linspace(1,5,6)
moderate=np.linspace(2,5,6)
black=np.linspace(2,6,6)

# print(white,moderate,black)

me=np.mean(white)
# print(me)
s=np.std(white)
x=fuzzy.gaussmf(white,me+0.4,s+0.4)

plt.plot(white,x)
# plt.show()

me1=np.mean(moderate)
s1=np.std(moderate)
x1=fuzzy.gaussmf(moderate,me1+0.4,s1+0.5)

me2=np.mean(black)
s2=np.std(black)
x2=fuzzy.gaussmf(black,me2+0.5,s2+0.5)

plt.plot(moderate,x1)
plt.plot(black,x2)
plt.show()
plt.legend()

```
