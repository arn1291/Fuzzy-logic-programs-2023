```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzzy


temp=np.linspace(1,40,50)

warm_membership=fuzzy.trapmf(temp,[32,33,34,35])
moderate_membership=fuzzy.trapmf(temp,[25,26,27,28])
cold_membership=fuzzy.trapmf(temp,[1,2,3,4])
hot_membership=fuzzy.trapmf(temp,[35,36,37,40])

plt.plot(temp, warm_membership, label="warm")
plt.plot(temp, moderate_membership, label="moderate")
plt.plot(temp, cold_membership, label="cold")
plt.plot(temp, hot_membership, label="hot")
plt.legend()
plt.show()
```
