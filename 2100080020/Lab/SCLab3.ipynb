```python
#In lab-2

import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
race = np.linspace(1, 10, 100)
wm = fuzz.gaussmf(race, 1.75, 1)
mm = fuzz.gaussmf(race, 4.75, 1)
bm = fuzz.gaussmf(race, 7.75, 1)
plt.plot(race, wm, label='White')
plt.plot(race, mm, label='Moderate')
plt.plot(race, bm, label='Black')
plt.legend()
plt.show()
```

```python

#Post lab

import matplotlib.pyplot as plt
import numpy as np
import skfuzzy as fuzz
temp = np.linspace(10, 80, 100)
wm = fuzz.trapmf(temp, [40, 55, 60, 65])
mm = fuzz.trapmf(temp, [30, 35, 40, 45])
cm = fuzz.trapmf(temp, [-10, -5, 0, 5])
hm = fuzz.trapmf(temp, [50, 55, 60, 65])
plt.plot(temp, wm, label='Warm')
plt.plot(temp, mm, label='Moderate')
plt.plot(temp, cm, label='Cold')
plt.plot(temp, hm, label='Hot')
plt.legend()
plt.show()
```

```python

# Skill

import numpy as np
import matplotlib.pyplot as plt
pts=np.array([[1,1],[2,2],[5,5],[6,6]])
data=np.array([[0.2,0.3,0.4,0.5],[0.8,0.7,0.6,0.5]])
noc=2
def centroids(pts,data,noc):
    arr=np.array([])
    dim=np.shape(pts)
    for i in range(0,noc):
        for j in range(0,dim[1]):
            dpts=pts[: ,j]
            #print(dpts)
            mem=data[i]**2
            #print(mem)
            num=np.dot(dpts,mem)
            s=np.sum(num)
            #div=num/s
            arr=np.append(arr,s)
            #arr.append(num)
            #print(dpts,mem,num)
    print(arr)
    return arr.reshape(noc,dpts[1])
a=centroids(pts,data,noc)
print(a)
def eucli(a,pts):
    mat=np.zeros([len(pts),len(a)])
    for i in range(len(pts)):
        for j in range(len(a)):
            mat[i][j]=np.linalg.norm(pts[i]-a[j])
    return mat
b=eucli(a,pts)
print(b)
```
