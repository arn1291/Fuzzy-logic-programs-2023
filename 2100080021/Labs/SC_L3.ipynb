```python
import matplotlib.pyplot as plt
import skfuzzy as fuzz
import numpy as np



height=np.linspace(3,6.5,35)

tall_mf = fuzz.trimf(height, [5.8, 6.2, 6.7])
verytall_mf = fuzz.trimf(height, [6.1, 6.4, 6.8])
normal_mf = fuzz.trimf(height, [5.2, 5.7, 6.0])
veryshort_mf = fuzz.trimf(height, [3.0, 3.5, 4.0])
short_mf = fuzz.trimf(height, [4.2, 4.7, 5.2])


plt.plot(height,normal_mf, label='Normal')
plt.plot(height,tall_mf ,label='Tall')
plt.plot(height,verytall_mf ,label='Very Tall')
plt.plot(height,veryshort_mf, label='Very Short')
plt.plot(height,short_mf, label='Short')


plt.legend()
plt.show()
```

```python
import numpy as np
import skfuzzy as fuzzy
import matplotlib.pyplot as plt

white=np.linspace(1,5,6)
moderate=np.linspace(2,5,6)
black=np.linspace(2,6,6)

me=np.mean(white)

s=np.std(white)
x=fuzzy.gaussmf(white,me+0.4,s+0.4)

plt.plot(white,x)

me1=np.mean(moderate)
s1=np.std(moderate)
x1=fuzzy.gaussmf(moderate,me1+0.4,s1+0.5)

me2=np.mean(black)
s2=np.std(black)
x2=fuzzy.gaussmf(black,me2+0.5,s2+0.5)

plt.plot(moderate,x1)
plt.plot(black,x2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzzy


temp = np.linspace(1, 40, 50)

warm_membership = fuzzy.trapmf(temp, [28, 32, 34, 36])
moderate_membership = fuzzy.trapmf(temp, [22, 25, 27, 30])
cold_membership = fuzzy.trapmf(temp, [-5, 0, 4, 8])
hot_membership = fuzzy.trapmf(temp, [36, 37, 38, 40])


plt.plot(temp, warm_membership, label="warm")
plt.plot(temp, moderate_membership, label="moderate")
plt.plot(temp, cold_membership, label="cold")
plt.plot(temp, hot_membership, label="hot")
plt.legend()
plt.show()
```
